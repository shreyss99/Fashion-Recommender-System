```python
def copy_images(source_folder, destination_folder):
    # Create the destination folder if it doesn't exist
    if not os.path.exists(destination_folder):
        os.mkdir(destination_folder)

    # Loop through each subfolder and its files
    for root, _, files in os.walk(source_folder):
        for file in files:
            if file.endswith(('.png', '.jpg', '.jpeg')):
                source_path = os.path.join(root, file)
                destination_path = os.path.join(destination_folder, file)
                shutil.copy(source_path, destination_path)
                print(f"Copied {file} to {destination_folder}")
            else:
                print(f"File {file} does not have a .png, .jpg or jpeg extension")

    print("All images have been copied successfully.")

# Example usage
source_folder = TRAINING_PATH
destination_folder = DESTINATION_PATH

copy_images(source_folder, destination_folder)
```

```python
RESIZED_IMAGES_DESTINATION_PATH = '/Users/shrey_shah/Desktop/Fashion/dataset/Raw_Images_Resized'
```

```python
def resize_images(source_folder, output_folder, size):
    """
    Resizes all images to specified size from the source folder and saves them in the output folder.
    """
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    # Loop through each image file in the source folder
    for file in os.listdir(source_folder):
        file_path = os.path.join(source_folder, file)
        
        # Read the image using OpenCV
        image = cv2.imread(file_path)
        
        # Resize the image
        resized_image = cv2.resize(image, size)
        
        # Save the resized image to the output folder
        output_path = os.path.join(output_folder, file)
        cv2.imwrite(output_path, resized_image)
        
        print(f"Resized {file} and saved to {output_folder}")
    
    print("All images have been resized and saved successfully.")
    
# Example usage
source_folder = RAW_IMAGES_DESTINATION_PATH
destination_folder = RESIZED_IMAGES_DESTINATION_PATH
size = (800, 600)  # Specify the desired width and height

resize_images(source_folder, destination_folder, size)
```

```python
AUGMENTED_IMAGES_DESTINATION_PATH = '/Users/shrey_shah/Desktop/Fashion/dataset/Augmented_Images'
```

```python
def augment_images(source_folder, output_folder):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Define the augmentation transformations
    augmentations = [
        A.Perspective(p=1.0, scale=(0.05, 0.15)),
        A.ElasticTransform(p=1.0, alpha=36, sigma=6),
        A.Rotate(p=1.0, limit=30, interpolation=cv2.INTER_LINEAR, border_mode=cv2.BORDER_REFLECT_101),
        #A.Shear(p=1.0, shear_limit=0.3, interpolation=cv2.INTER_LINEAR, border_mode=cv2.BORDER_REFLECT_101),
        A.Crop(p=1.0, x_min=100, y_min=100, x_max=300, y_max=300),
        A.HorizontalFlip(p=1.0)
    ]

    # Loop through each image file in the source folder
    for file_name in os.listdir(source_folder):
        file_path = os.path.join(source_folder, file_name)

        # Read the image using OpenCV
        image = cv2.imread(file_path)

        # Apply each augmentation transformation
        for i, augmentation in enumerate(augmentations):
            augmented = augmentation(image=image)
            augmented_image = augmented["image"]

            # Save the augmented image to the output folder
            output_path = os.path.join(output_folder, f"augmented_{i}_{file_name}")
            cv2.imwrite(output_path, augmented_image)

    print("All images have been augmented and saved successfully.")

# Example usage
source_folder = RESIZED_IMAGES_DESTINATION_PATH
output_folder = AUGMENTED_IMAGES_DESTINATION_PATH

augment_images(source_folder, output_folder)
```

```python
def gray_scale_images(source_folder, output_folder):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    # Define the grayscale transformation
    transform = A.Compose([
        A.Lambda(image=lambda img: cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))
    ])
    
    # Loop through each image file in the source folder
    for file_name in os.listdir(source_folder):
        file_path = os.path.join(source_folder, file_name)

        # Read the image using OpenCV
        image = cv2.imread(file_path)

        # Apply the grayscale transformation
        grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        # Save the grayscale image to the output folder
        output_path = os.path.join(output_folder, f"grayscale_{file_name}")
        cv2.imwrite(output_path, grayscale_image)

    print("All images have been converted to grayscale and saved successfully.")


# Example usage
source_folder = AUGMENTED_IMAGES_DESTINATION_PATH
output_folder = AUGMENTED_GRAYSCALE_IMAGES_DESTINATION_PATH

gray_scale_images(source_folder, output_folder)
```
